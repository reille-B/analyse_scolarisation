```python
# Analyse descriptive : scolarisation des filles en Afrique centrale 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Paramètres généraux

# Liste des codes pays de la CEMAC + pays voisins
pays_codes = ["BDI", "CMR", "CAF", "TCD", "COG", "COD", "GNQ", "GAB", "RWA", "STP"]
pays_noms = {
    "BDI": "Burundi",
    "CMR": "Cameroun",
    "CAF": "Centrafrique",
    "TCD": "Tchad",
    "COG": "Congo",
    "COD": "RD Congo",
    "GNQ": "Guinee eq.",
    "GAB": "Gabon",
    "RWA": "Rwanda",
    "STP": "Sao Tome"
}

# Precision de données et de sortie
dossier_data = "../data"
dossier_out = "../figures"
```

```python
# Création du dossier de sortie si il n'existe pas
os.makedirs(dossier_out, exist_ok=True)


# Fonction de chargement des fichiers UNESCO

def charger_unesco(nom_dossier):
    # Construction du chemin vers le fichier CSV
    chemin = os.path.join(dossier_data, nom_dossier, "data.csv")

    # Lecture du fichier
    df = pd.read_csv(chemin)

    # On garde uniquement les pays qui nous intéressent
    df = df[df["geoUnit"].isin(pays_codes)]

    # On garde uniquement la période 2000-2024
    df = df[(df["year"] >= 2000) & (df["year"] <= 2024)]

    # Ajout de la colonne avec le nom du pays
    df["pays"] = df["geoUnit"].map(pays_noms)

    return df

# Chargement des données

# Taux brut de scolarisation (GER) - primaire et supérieur
ger_prim = charger_unesco("indicator-data-export_GER.1.F_GER.1.GPIA_GER.1.M_and_3_more")
ger_sup  = charger_unesco("indicator-data-export_GER.3.F_GER.3.GPIA_GER.3.M")

# Taux d'achèvement scolaire
completion = charger_unesco("indicator-data-export_CR.MOD.1.F_CR.MOD.1.M_CR.MOD.2.F_and_3_more")

# On combine primaire et supérieur dans un seul tableau
ger_all = pd.concat([ger_prim, ger_sup], ignore_index=True)

print("Nombre d'observations GER :", len(ger_all))
print("Nombre d'observations Completion :", len(completion))
```

```python
# FIGURE 1 : Taux brut de scolarisation - FILLES
#             par niveau d'enseignement

# On filtre uniquement les indicateurs pour les filles
ger_filles = ger_all[ger_all["indicatorId"].isin(["GER.1.F", "GER.2.F", "GER.3.F"])].copy()

# On crée une colonne lisible pour le niveau scolaire
ger_filles["niveau"] = ger_filles["indicatorId"].map({
    "GER.1.F": "Primaire",
    "GER.2.F": "Secondaire inf.",
    "GER.3.F": "Secondaire sup."
})

# Création de la figure avec 3 sous-graphiques côte à côte
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

niveaux = ["Primaire", "Secondaire inf.", "Secondaire sup."]

for i, niv in enumerate(niveaux):
    ax = axes[i]

    # Données pour ce niveau uniquement
    data_niv = ger_filles[ger_filles["niveau"] == niv]

    # Une courbe par pays
    for pays in data_niv["pays"].unique():
        d = data_niv[data_niv["pays"] == pays]
        ax.plot(d["year"], d["value"], marker="o", markersize=3, label=pays)

    ax.set_title(niv)
    ax.set_xlabel("Année")
    ax.set_ylabel("TBS filles (%)")
    ax.legend(fontsize=6)
    ax.grid(alpha=0.3)

plt.suptitle("Fig 1 — Taux brut de scolarisation filles, Afrique centrale (2000-2024)",
             fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig1_ger_filles.png"), dpi=300)
plt.show()
print("Figure 1 sauvegardée")
```

```python
# FIGURE 2 : Indice de parité de genre (IPG)

# On filtre les indicateurs de parité (GPIA = Gender Parity Index Adjusted)
gpi = ger_all[ger_all["indicatorId"].isin(["GER.1.GPIA", "GER.2.GPIA", "GER.3.GPIA"])].copy()

# Nom lisible pour chaque niveau
gpi["niveau"] = gpi["indicatorId"].map({
    "GER.1.GPIA": "Primaire",
    "GER.2.GPIA": "Secondaire inf.",
    "GER.3.GPIA": "Secondaire sup."
})

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, niv in enumerate(["Primaire", "Secondaire inf.", "Secondaire sup."]):
    ax = axes[i]
    data_niv = gpi[gpi["niveau"] == niv]

    for pays in data_niv["pays"].unique():
        d = data_niv[data_niv["pays"] == pays]
        ax.plot(d["year"], d["value"], marker="o", markersize=3, label=pays)

    # Ligne de référence : IPG = 1 signifie la parité parfaite
    ax.axhline(1, color="black", linestyle="--", alpha=0.5, label="Parité")

    ax.set_title(niv)
    ax.set_xlabel("Année")
    ax.set_ylabel("IPG")
    ax.legend(fontsize=6)
    ax.grid(alpha=0.3)

plt.suptitle("Fig 2 — Indice de parité de genre (IPG)", fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig2_ipg.png"), dpi=300)
plt.show()
print("Figure 2 sauvegardée")
```

```python
# FIGURE 3 : Cascade d'achèvement scolaire - FILLES
#             Moyenne sur la période récente (2018-2024)

# On garde uniquement les indicateurs d'achèvement pour les filles
comp_filles = completion[completion["indicatorId"].isin(
    ["CR.MOD.1.F", "CR.MOD.2.F", "CR.MOD.3.F"])].copy()

# Nom lisible pour chaque niveau
comp_filles["niveau"] = comp_filles["indicatorId"].map({
    "CR.MOD.1.F": "Primaire",
    "CR.MOD.2.F": "Sec. inf.",
    "CR.MOD.3.F": "Sec. sup."
})

# On calcule la moyenne par pays et par niveau sur 2018-2024
comp_recente = comp_filles[comp_filles["year"] >= 2018]
comp_agg = comp_recente.groupby(["pays", "niveau"])["value"].mean().reset_index()

# On réorganise le tableau pour avoir les niveaux en colonnes
comp_pivot = comp_agg.pivot(index="pays", columns="niveau", values="value")
comp_pivot = comp_pivot[["Primaire", "Sec. inf.", "Sec. sup."]]

# Graphique en barres groupées
fig, ax = plt.subplots(figsize=(12, 6))
comp_pivot.plot(kind="bar", ax=ax, colormap="Blues_r", width=0.75)

ax.set_title("Fig 3 — Cascade d'achèvement scolaire filles (~2018-2024)", fontweight="bold")
ax.set_ylabel("Taux d'achèvement (%)")
ax.set_ylim(0, 100)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig3_cascade.png"), dpi=300)
plt.show()
print("Figure 3 sauvegardée")

congo_prim = comp_agg[(comp_agg['pays']=='Congo') & (comp_agg['niveau']=='Primaire')]['value'].values[0]
congo_sup  = comp_agg[(comp_agg['pays']=='Congo') & (comp_agg['niveau']=='Sec. sup.')]['value'].values[0]
print(f"Congo — Primaire: {congo_prim:.1f}%, Sec. sup.: {congo_sup:.1f}%")
```

```python
# FIGURE 4 : Écart de genre dans l'achèvement scolaire
#             (Taux filles - Taux garçons)

# On identifie le sexe à partir du code d'indicateur
ecart = completion.copy()

ecart["sexe"] = ecart["indicatorId"].apply(
    lambda x: "F" if x.endswith(".F") else ("M" if x.endswith(".M") else None)
)

ecart["niv"] = ecart["indicatorId"].apply(
    lambda x: x.split(".")[2] if len(x.split(".")) > 2 else None
)

# On supprime les lignes sans information de sexe ou de niveau
ecart = ecart.dropna(subset=["sexe", "niv"])

# On pivote pour avoir F et M en colonnes
pivot = ecart.pivot_table(
    index=["pays", "year", "niv"],
    columns="sexe",
    values="value"
).reset_index()

pivot = pivot.dropna()

# Calcul de l'écart : positif = filles devant, négatif = garçons devant
pivot["ecart"] = pivot["F"] - pivot["M"]

# Tracé
fig, ax = plt.subplots(figsize=(11, 5))

for pays in pivot["pays"].unique():
    d = pivot[pivot["pays"] == pays]
    ax.plot(d["year"], d["ecart"], marker="o", markersize=3, label=pays)

# Ligne zéro = égalité parfaite
ax.axhline(0, color="black", linestyle="--", alpha=0.5)

ax.set_title("Fig 4 — Écart de genre dans l'achèvement scolaire (Filles - Garçons)",
             fontweight="bold")
ax.set_xlabel("Année")
ax.set_ylabel("Écart (points de pourcentage)")
ax.legend(fontsize=7, ncol=2)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig4_ecart_genre.png"), dpi=300)
plt.show()
print("Figure 4 sauvegardée")
```

```python
# Chargement du Taux Net de Scolarisation (TNS)

chemin_nert = os.path.join(
    dossier_data,
    "indicator-data-export_NERT.1.F.CP_NERT.1.M.CP_NERT.2.F.CP_and_3_more",
    "data.csv"
)

nert = pd.read_csv(chemin_nert)
nert = nert[nert["geoUnit"].isin(pays_codes)]
nert = nert[(nert["year"] >= 2000) & (nert["year"] <= 2024)]
nert["pays"] = nert["geoUnit"].map(pays_noms)

print("Nombre d'observations TNS :", len(nert))
print("Indicateurs disponibles :", nert["indicatorId"].unique())
```

```python
# FIGURE 5 : Taux Net de Scolarisation - FILLES

# Filtrage des indicateurs TNS pour les filles
tns_filles = nert[nert["indicatorId"].isin(
    ["NERT.1.F.CP", "NERT.2.F.CP", "NERT.3.F.CP"])].copy()

tns_filles["niveau"] = tns_filles["indicatorId"].map({
    "NERT.1.F.CP": "Primaire",
    "NERT.2.F.CP": "Secondaire inf.",
    "NERT.3.F.CP": "Secondaire sup."
})

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, niv in enumerate(["Primaire", "Secondaire inf.", "Secondaire sup."]):
    ax = axes[i]
    data_niv = tns_filles[tns_filles["niveau"] == niv]

    for pays in data_niv["pays"].unique():
        d = data_niv[data_niv["pays"] == pays]
        ax.plot(d["year"], d["value"], marker="o", markersize=3, label=pays)

    ax.set_title(niv)
    ax.set_xlabel("Année")
    ax.set_ylabel("TNS filles (%)")
    ax.legend(fontsize=6)
    ax.grid(alpha=0.3)

plt.suptitle("Fig 5 — Taux Net de Scolarisation filles, Afrique centrale (2000-2024)",
             fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig5_tns_filles.png"), dpi=300)
plt.show()
print("Figure 5 sauvegardée")
```

```python
# Chargement des enfants hors école (OOS)

chemin_oos = os.path.join(dossier_data, "OOS_Rate_Countries.csv")

oos = pd.read_csv(chemin_oos)
oos = oos[oos["country"].isin(pays_codes)]
oos = oos[(oos["year"] >= 2000) & (oos["year"] <= 2024)]
oos["pays"] = oos["country"].map(pays_noms)

print("Nombre d'observations OOS :", len(oos))
print("Niveaux disponibles :", oos["level"].unique())
print("Sexes disponibles :", oos["sex"].unique())
```

```python
# FIGURE 6 : Taux d'enfants hors école - FILLES

# On garde uniquement les filles
oos_filles = oos[oos["sex"] == "female"].copy()
niveaux_oos = sorted(oos_filles["level"].unique())

fig, axes = plt.subplots(1, len(niveaux_oos), figsize=(15, 5))

for i, niv in enumerate(niveaux_oos):
    ax = axes[i]
    data_niv = oos_filles[oos_filles["level"] == niv]

    for pays in data_niv["pays"].unique():
        d = data_niv[data_niv["pays"] == pays]
        ax.plot(d["year"], d["value"], marker="o", markersize=3, label=pays)

    ax.set_title(niv)
    ax.set_xlabel("Année")
    ax.set_ylabel("Taux hors école (%)")
    ax.legend(fontsize=6)
    ax.grid(alpha=0.3)

plt.suptitle("Fig 6 — Taux d'enfants hors école - Filles", fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(dossier_out, "fig6_oos_filles.png"), dpi=300)
plt.show()
print("Figure 6 sauvegardée")
```
